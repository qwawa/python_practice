```python
import openai
import csv
import openpyxl
```

```python
def summarize_text_korean(text):
    if text == "":
        return "NAN"
    response = openai.Completion.create(
        engine="text-davinci-003",
        prompt=f"이 한국어 텍스트를 요약해줘: {text}",
        max_tokens=500,
        temperature=0.3,
        top_p=1.0,
        frequency_penalty=0.0,
        presence_penalty=0.0
    )
    summary_text = response.choices[0].text.strip()
    return summary_text
```

```python
def extract_key_sentences_korean(text, num_sentences=3):
    if text == "":
        return "NAN"
    response = openai.Completion.create(
        engine="text-davinci-003",
        prompt=f"이 한국어 텍스트에서 핵심 문장 3개를 만들어줘: {text}",
        max_tokens=400,
        temperature=0.3,
        top_p=1.0,
        frequency_penalty=0.0,
        presence_penalty=0.0
    )
    summary_text = response.choices[0].text.strip()
    return summary_text
```

```python
# 텍스트 읽기 함수
def read_csv_column(file_path, column_index):
    strings = []
    with open(file_path, 'r', encoding='utf-8') as file:
        reader = csv.reader(file)
        for row in reader:
            if len(row) > column_index:
                strings.append(row[column_index])
    return strings
```
